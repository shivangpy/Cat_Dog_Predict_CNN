```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import PIL
```

```python
CATEGORIES=['Dog','Cat']
```

```python
image=r'C:\Users\This PC\Downloads\kagglecatsanddogs_3367a\Test_images\40.jpg'
```

```python
def prepare(filepath):
    img_size=100
    img_array=cv2.imread(image,cv2.IMREAD_GRAYSCALE)
    new_array=cv2.resize(img_array,(img_size,img_size))
    return new_array.reshape(-1,img_size,img_size,1)
```

```python
model=tf.keras.models.load_model(r'C:\Users\This PC\Downloads\kagglecatsanddogs_3367a\Dogs_vs_Cats_CNN.model')
prediction=model.predict([prepare(image)])
print(CATEGORIES[int(prediction[0][0])])
img=mpimg.imread(image)
imgplot=plt.imshow(img)
plt.title(CATEGORIES[int(prediction[0][0])])
```
